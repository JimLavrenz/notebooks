# Gradient Descent (Deep Learning Tutorial)

In this tutorial we computer the minimum of a simple multivariate function 

## $$f(x_1, x_2) = x_1^2 + bx_2^2$$ 

using gradient descent. Where b is chosen to a elliptic countour plot to show the jaggedness of using a first order gradient descent algorithm. We use a simple fixed learning rate $s$ (sometimes called $\lambda$)

To compute the min, we use the gradient descent algorithm,

## $$x_i^{k+1} = x_i^k - s\nabla f^k(x_i^k)$$

Where i is the feature and k is the iteration as we step down the gradient. For the case we are showing, you may think of $x_1$ as simply x and $x_2$ as simply y and combined they'd form a column vector. It should be understood that the superscript k is not taking the power of anything rather just serves as another index.

Recall that the gradient is a vector itself and is equal to 

$$\frac{\partial}{\partial x_1} \hat{x_1} + \frac{\partial}{\partial x_2} \hat{x_2}$$

So for our simple convex function we know it will have a minimum at ($x_1$ = 0, $x_2$ = 0) since it's an ellipse in countour shape and parabolic overall.

$$\nabla f = 2x_1 \hat{x_1} + 2x_2 \hat{x_2}$$
  
Note for this problem we are using the simple 1st order gradient descent algorithm. Others exist that use acceleration (second order terms). And still others that use stochastic methods for large training data sets.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
b = 0.5

def f(x_1, x_2):
    return x_1**2 + b * x_2**2
def dx_1(x_1):
    return 2 * x_1
def dx_2(x_2):
    return 2 * b * x_2
```

```python
xg = np.linspace(-2, 2, 100)
yg = np.linspace(-2, 2, 100)

X, Y = np.meshgrid(xg, yg)
Z = f(X, Y)
plt.contour(X, Y, Z, colors='black');
plt.title("Contour plot of f(x_1, x_2) with b = 0.5")
plt.xlabel("x_1 axis")
plt.ylabel("x_2 axis")
plt.savefig('gd_img_1.png', dpi=300)
```

```python
#step size
s = 0.30

#number of iterations
k_max = 20

init_x_1 = 2
init_x_2 = 2

x_1 = np.array([])
x_2 = np.array([])

x_1 = np.append(x_1, init_x_1)
x_2 = np.append(x_2, init_x_2)

#iterations
k = 0

while k < k_max:
    x_1 = np.append(x_1, x_1[k] - s * dx_1(x_1[k]))
    x_2 = np.append(x_2, x_2[k] - s * dx_2(x_2[k]))
    k += 1
```

```python
print("Final error from (0,0): x_1 = ", x_1[k])
print("Final error from (0,0): x_2 = ", x_2[k])
```

```python
plt.plot(x_1, x_2)
plt.title('x_1_k vs x_2_k plotted from (x_1_0, x_2_0)')
plt.savefig('gd_img_2.png', dpi=300)
```

```python
x_axis = np.linspace(0, k_max, k_max + 1)
plt.plot(x_axis, x_1)
plt.title("x_1 distance from min")
plt.xlabel("k iterations")
plt.ylabel("x_1 distance from min")
plt.savefig('gd_img_3.png', dpi=300)
```

```python
plt.plot(x_axis, x_2)
plt.title("x_2 distance from min")
plt.xlabel("k iterations")
plt.ylabel("x_2 distance from min")
plt.savefig('gd_img_4.png', dpi=300)
```
